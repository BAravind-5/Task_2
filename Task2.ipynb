```python
Df2=pd.read_csv("products_missing_demo.csv")
```

```python
Df2.isnull().sum()
```

Missing Value Observation:

The dataset contains a small number of missing values across both numerical and
categorical columns. This is a realistic scenario commonly found in real-world data.

```python
Df2.isnull().sum().plot(kind='bar', title='Missing Values per Column')
```

The bar chart shows that missing values are limited and evenly distributed.

```python
Df2['Price'] = Df2['Price'].fillna(Df2['Price'].median())
Df2['Units_Sold'] = Df2['Units_Sold'].fillna(Df2['Units_Sold'].mean())
Df2['Revenue'] = Df2['Revenue'].fillna(Df2['Revenue'].mean())
Df2['Trend_Score'] = Df2['Trend_Score'].fillna(Df2['Trend_Score'].mean())
```

```python
Df2.isnull().sum()
```

```python
Df2['Subcategory'] = Df2['Subcategory'].fillna(Df2['Subcategory'].mode()[0])
```

```python
Df2.isnull().sum()
```

```python
print("Before cleaning: Missing values present")
print("After cleaning: No missing values")
print("Dataset shape:", Df2.shape)
```

After imputation, all missing values were handled without removing any rows,
preserving dataset size and improving data quality.
